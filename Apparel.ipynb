```python
import pickle
import csv

class Apparel:
    """
    Represents an apparel item.

    Attributes:
        date (str): Date on which the apparel item is added to the website.
        name (str): Name of the apparel.
        category (str): Category of the apparel.
        type (str): Type of apparel.
        rating (float): Customers's rating for the apparel.
        price (float): Price of the apparel item (in USD).
        production_cost (float): Cost incurred to manufacture the apparel item (in USD).
        unique_id (int): Unique Identifier.
    """
    def __init__(self, date, name, category, _type, rating, price, production_cost):
        self.date = date
        self.name = name
        self.category = category
        self.type = _type
        self.rating = rating
        self.price = price
        self.production_cost = production_cost
        self.unique_id = id(self)

    def __str__(self):
        return f"{self.unique_id},{self.date},{self.name},{self.category},{self.type},{self.rating},{self.price},{self.production_cost}"

class Pants(Apparel):
    """
    Represents a pants apparel item.

    Inherits from Apparel class.
    """
    def __init__(self, date, name, category, rating, price, production_cost):
        super().__init__(date, name, category, "Pants", rating, price, production_cost)

class Shirt(Apparel):
    """
    Represents a shirt apparel item.

    Inherits from Apparel class.
    """
    def __init__(self, date, name, category, rating, price, production_cost):
        super().__init__(date, name, category, "Shirt", rating, price, production_cost)

class Jeans(Pants):
    """
    Represents a jeans apparel item.

    Inherits from Pants class.
    """
    def __init__(self, date, name, rating, price, production_cost):
        super().__init__(date, name, "Pants", rating, price, production_cost)

class TShirt(Shirt):
    """
    Represents a T-shirt apparel item.

    Inherits from Shirt class.
    """
    def __init__(self, date, name, rating, price, production_cost):
        super().__init__(date, name, "Shirt", rating, price, production_cost)

class FormalShirt(Shirt):
    """
    Represents a formal shirt apparel item.

    Inherits from Shirt class.
    """
    def __init__(self, date, name, rating, price, production_cost):
        super().__init__(date, name, "Shirt", rating, price, production_cost)

# Testing the code
PantsApparel = Pants("2022-03-13", "Distressed Jeans", "Jeans", 3.0, 69.94, 42.89)
print(str(PantsApparel))

# Loading the data from pickle file
with open('Apparelpkl638200303.dat', 'rb') as f:
    apparel_data = pickle.load(f)

# Writing data to CSV file
with open('apparel_data.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["unique_id", "date", "name", "category", "type", "rating", "price", "production_cost"])
    for item in apparel_data:
        writer.writerow(str(item).split(','))  # Assuming item is an instance of Apparel or its subclasses
```

```python
# For categorical 'rating', use mode
rating_mode = app_data['rating'].mode()[0]
app_data['rating'].fillna(rating_mode, inplace=True)

# For numerical 'price', use median
cost_median = app_data['price'].median()
app_data['price'].fillna(cost_median, inplace=True)

# For numerical 'production_cost', use median
production_cost_median = app_data['production_cost'].median()
app_data['production_cost'].fillna(production_cost_median, inplace=True)

# Convert 'date' to datetime format (for line plot)
app_data['date'] = pd.to_datetime(app_data['date'])
ad_data_sorted = app_data.sort_values('date')
```

```python
#counting number of instances in each category
app_data['category'].value_counts()
```

```python
y=app_data['category'].value_counts()
y
```

```python
# 1. Bar Chart: Frequency of Apparel by category

plt.bar(y.index,y.values,color='blue')
plt.title('Frequency of Apparel by category')
plt.xlabel('category')
plt.ylabel('Frequency')
plt.show()
```

```python
#bar chart graph Insights:

#The bar plot depicts the frequency of different apparel categories: Jeans, T-Shirts, and Formal Shirts.
#Jeans and T-Shirts exhibit significantly higher frequencies, both approaching 5000. ​
#In contrast, Formal Shirts have a much lower frequency, below 1000. ​
#This data suggests that Jeans and T-Shirts are more commonly purchased compared to Formal Shirts. ​
```

```python
# 2. Boxplot: Distribution of Price Across Different Category
plt.figure(figsize=(12, 8))
sns.boxplot(x='category', y='price', data=app_data)
plt.title('Distribution of Price Across Different category')
plt.xlabel('category')
plt.ylabel('price')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

```python
#Box plot Insights:

#This box plot shows us the selling price distribution of 3 categories of apparel.​
#We can clearly see that much of the price point of all 3 categories lies between 60 and 80 with some outliers.​
```

```python
# 3. Line Plot: Trends in Price Over Time
plt.figure(figsize=(12, 6))
sns.lineplot(x='date', y='price', data=app_data)
plt.title('Trends in price Over Time')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```

```python
#Insighs from Line plot

#There is no clear upward or downward trend. ​
#There are fluctuations within the data that could suggest any number of other factors. ​
#The plot is right skewed.​
```

```python
# 4. Scatter Plot: Relationship Between Price and Production Cost
plt.figure(figsize=(40, 10))
sns.scatterplot(x='price', y='production_cost', data=app_data)
plt.title('Relationship Between Price and Production Cost')
plt.xlabel('Price')
plt.ylabel('Production cost')
plt.show()
```

```python
#insights from Scatter plot

#Min cost of production of apparel is 47.5 with max cost of production is 57.5​
#Most of the apparel are being sold at the price range of 60 to 80 with some exceptions.​
#Based on the data from this scatter plot we can assume that this is profit making.​
```

```python
# 5. Histogram: Distribution of Production cost
plt.figure(figsize=(10, 6))
sns.histplot(app_data['production_cost'], bins=30)
plt.title('Distribution of Production cost')
plt.xlabel('Production cost')
plt.ylabel('Frequency')
plt.show()
```

```python
#Insights from Histogram

#From the graph we can conclude that the most frequent production cost is 44 with close to 1000 items and the least frequent production cost is 57.5 with less than 100 items. ​
#The production cost for most of the items is between 40 and 45. ​
```

```python
# 6. Pie Chart: Proportion of Apparel Category
type_counts = app_data['category'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Proportion of Apparel category')
plt.show()
```

```python
# 7. Heat Map: Correlation Between Numerical Variables
app_data['date_bin'] = app_data['date'].dt.to_period('M')

# Create a pivot table with 'date_bin' as rows and 'tag' as columns
pivot_table = app_data.pivot_table(index='date_bin', columns='category', aggfunc='size', fill_value=0)

# Generate the heat map
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, cmap='viridis')
plt.title('Heat Map of category Frequency Over Time')
plt.ylabel('Date')
plt.xlabel('category')
plt.show()
```

```python
#Heat Map Insights

#It covers data from January 2022 (2022-01) to March 2023 (2023-01).​
#Darker colors indicate higher frequencies.​
#Lighter colors correspond to lower frequencies.​
#In January 2022, Formal Shirts had a high frequency (dark color).​
#T-Shirts had a lower frequency during the same period (lighter color).​
#Jeans maintained a moderate frequency across most months.​
```
